```python
crop_data = []

for crop in crops:
    temp_range = crop['temperature'].split('-')
    humidity_range = crop['humidity'].split('-')
    ph_range = crop['ph'].split('-')
    rainfall_range = crop['rainfall'].split('-')
    
    if len(humidity_range) == 1:
        humidity_range.append(humidity_range[0])
    
    if len(ph_range) == 1:
        ph_range.append(ph_range[0])
    
    if len(rainfall_range) == 1:
        rainfall_range.append(rainfall_range[0])
    
    temperatures = np.random.uniform(float(temp_range[0]), float(temp_range[1]), 150)
    humidity = np.random.uniform(float(humidity_range[0]), float(humidity_range[1]), 150)
    ph = np.random.uniform(float(ph_range[0]), float(ph_range[1]), 150)
    rainfall = np.random.uniform(float(rainfall_range[0]), float(rainfall_range[1]), 150)
    
    for i in range(0,150):
        crop_data.append({
            'temperature': round(temperatures[i], 3),
            'humidity': round(humidity[i], 3),
            'ph': round(ph[i], 3),
            'rainfall': round(rainfall[i], 3),
            'label': crop['label']
        })
        
df = pd.DataFrame(crop_data, columns=['temperature', 'humidity', 'ph', 'rainfall', 'label'])
df
```
